```python
import openai
import os

def paraphrase(query):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo-0613",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": "Give me 4 paraphrased sentences of  \""+ query},
        ]
    )

    answer = response.choices[0].message.content
    return answer
```

```python
# Open and load the caption5.json file
filename = "caption5.json"
with open(filename, "r") as file:
    data = json.load(file)
print(len(data))
```

```python
import re
pattern = r'[A-Z][^.!?]*[.!?]'
```

```python
from time import sleep
```

```python
for idx, item in enumerate(data):
    if idx !=0 and idx %50 == 0:
      print("sleep---")
      sleep(20)

    sentence = item["label"]
    sentences = [sentence]

    paraphrased_sentences = paraphrase(sentence)

    # 정규식을 사용하여 문장 추출
    paraphrased_sentences = re.findall(pattern, paraphrased_sentences)

    for s in paraphrased_sentences:
      sentences.append(s.strip())

    item["label"] = sentences

    print(idx, item)
```
