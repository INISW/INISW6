```python
import json
from pycocotools.coco import COCO
from pycocoevalcap.eval import COCOEvalCap
```

```python
def coco_caption_eval(annotation_file, results_file):

    coco = COCO(annotation_file)
    coco_result = coco.loadRes(results_file)

    coco_eval = COCOEvalCap(coco, coco_result)
    coco_eval.evaluate()

    return coco_eval
```

```python
val = []
for i in range(30):
    val.append(coco_caption_eval("/content/drive/MyDrive/mjae/annotaions/gt_captions.json",f'/content/drive/MyDrive/mjae/output/val/{i+1}.json').eval.items())
```

```python
with open('/content//drive/MyDrive/mjae/ouput/score.json','w') as f:
  json.dump(val,f)
```
