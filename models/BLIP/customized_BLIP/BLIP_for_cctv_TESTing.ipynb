# caption 생성 체크

```python
from inference import single_image_caption, multi_image_caption
```

```python
import os

def get_file_paths(folder_path):
    file_paths = []
    for root, directories, files in os.walk(folder_path):
        for file in files:
            file_path = os.path.join(root, file)
            file_paths.append(file_path)
    return file_paths
```

```python
file_paths = get_file_paths('/content/drive/MyDrive/BLIP/export/datasets/coco/images/cctv100')
```

```python
from PIL import Image
import numpy as np

images = list()

for f in file_paths:
  im = Image.open(f)
  arry = np.asarray(im)
  im = Image.fromarray(arry)
  images.append(im)
```

```python
captions = multi_image_caption(images)
```
